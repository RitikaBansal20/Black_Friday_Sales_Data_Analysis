```python
import pandas as pd
df=pd.read_csv("BlackFriday.csv")
del df['Product_Category_2']
del df['Product_Category_3']
df
```

```python
df.groupby('Age').size()
```

```python
df.groupby('Age').size().plot(kind='bar'
                                 ,figsize =(12,6)
                                 ,title='Purchase Distribution by Age')
```

```python
for i in df['Age'].unique():
          print(i, " : ",df[df['Age']== i]['Product_ID'].nunique())
```

```python
lst=[]
for i in df['Age'].unique():
          lst.append([i,df[df['Age']== i]['Product_ID'].nunique()])
        
data=pd.DataFrame(lst, columns=['Age',"products"])
data
```

```python
data.plot.bar(x='Age', figsize=(8,6))
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind='bar'
                                 ,figsize =(12,6)
                                 ,title='Amount Spend by Age')
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind='pie'
                                 ,figsize =(12,6)
                                 ,autopct = "%.1f")
```
